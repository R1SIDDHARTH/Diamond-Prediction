```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# 1. Data Loading and Initial Exploration
data = pd.read_csv(r'C:\ALL folder in dexstop\PycharmProjects\dimond\archive (1)\diamonds.csv')
print(data.head())
print(data.info())
```

```python
# 2. Data Cleaning
# Drop the 'Unnamed: 0' column
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
# Removing any rows with dimensionless diamonds (x, y, z = 0)
data = data[(data['x'] > 0) & (data['y'] > 0) & (data['z'] > 0)]
```

```python
# Removing outliers based on reasonable ranges for depth and table
data = data[(data['depth'] < 75) & (data['depth'] > 45)]
data = data[(data['table'] < 80) & (data['table'] > 40)]
```

```python
# 3. Data Visualization (Optional but helpful for understanding)
palette = "viridis"
sns.pairplot(data, hue="cut", palette=palette)
plt.show()
```

```python
# Ensure that categorical columns are encoded to numeric values
label_encoder = LabelEncoder()
for col in ['cut', 'color', 'clarity']:
    data[col] = label_encoder.fit_transform(data[col])

# Now you can safely compute the correlation matrix
corrmat = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 12))
sns.heatmap(corrmat, annot=True, cmap="coolwarm")
plt.show()
```

```python
# 4. Data Encoding
# Encode categorical features
label_encoder = LabelEncoder()
for col in ['cut', 'color', 'clarity']:
    data[col] = label_encoder.fit_transform(data[col])
```

```python

# 5. Feature and Target Selection
X = data.drop('price', axis=1)  # Features
y = data['price']  # Target variable

# Split the data into training and testing sets (75% training, 25% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

```

```python

# 6. Model Training and Comparison
# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'XGBoost': XGBRegressor()
}

# Train models and store performance
results = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    results[name] = {'RMSE': rmse, 'R²': r2}

# Display results
for model_name, metrics in results.items():
    print(f"{model_name}: RMSE = {metrics['RMSE']:.2f}, R² = {metrics['R²']:.2f}")
```

```python
# 7. Feature Importance (For XGBoost)
xgb_model = models['XGBoost']
feature_importances = pd.Series(xgb_model.feature_importances_, index=X.columns)
plt.figure(figsize=(10, 6))
feature_importances.sort_values().plot(kind='barh', title='Feature Importances (XGBoost)')
plt.show()
```

```python
# 8. Visualization of Predictions (Actual vs Predicted)
best_model = models['XGBoost']  # You can choose the best model based on the results
y_pred_best = best_model.predict(X_test_scaled)

plt.scatter(y_test, y_pred_best, alpha=0.3)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Prices')
plt.show()
```
